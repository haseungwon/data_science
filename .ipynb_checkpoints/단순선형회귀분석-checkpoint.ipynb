### crim/rm/lstat 3개의 변수로 각각 단순 선형 회귀 분석

```python
## 변수 설정 target/crim/rm/lstat
target = boston[['Target']]
crim = boston[['CRIM']]
rm = boston[['RM']]
lstat = boston[['LSTAT']]
```

### target ~ crim 선형회귀분석

```python
# crim변수에 상수항 추가하기 y = b0+b1x 이기때문에
crim1 = sm.add_constant(crim, has_constant = "add") # Add a column of ones to an array.
```

```python
crim1.head()
```

```python
# sm.OLS 적합시키기 ordinary least squares
model1 = sm.OLS(target , crim1)
fitted_model1 = model1.fit()
```

```python
fitted_model1.summary()
# R-squared: ssr/sst
```

```python
# 회귀 계수 출력

# fitted_model1.params 의 반환은 series이다

print( fitted_model1.params)
print( fitted_model1.params['CRIM'])
```

### y_hat = beta0 + beta1*x 계산하기

```python
print(crim1.shape)
print(fitted_model1.params.shape)
```

```python
# 회귀 계수 x 데이터(X)
# crim(1 , x) * fitted_model1.params(const, 회귀계수)
np.dot(crim1, fitted_model1.params)
```

```python
# predict 함수를 통해 yhat 구하기
predict = fitted_model1.predict(crim1)
```

```python
## 직접 구한 yhat과 predict 함수를 통해 구한 yhat 차이
predict - np.dot(crim1, fitted_model1.params)
```

### 적합시킨 직선 시각화

```python
import matplotlib.pyplot as plt
plt.yticks(fontname = 'Arial')
plt.scatter(crim, target, label = 'data')
plt.plot(crim , predict, label = 'result')
plt.legend()
plt.show()
```

```python
plt.scatter(target, predict) 
plt.xlabel('real_value')
plt.ylabel('predict_value')
plt.show()
# 잘 맞지 않음
```

```python
## residual 시각화 잔차

fitted_model1.resid.plot()
plt.xlabel('resicual_number')
plt.show()
```

```python
# 잔차의 합계산하기 ->잔차는 0에 수렴할 것이다 
np.sum(fitted_model1.resid)
```

### rm/lstat 으로 해보기

```python
rm1 = sm.add_constant(rm, has_constant = "add") # Add a column of ones to an array.
lstat1 = sm.add_constant(lstat, has_constant = "add") # Add a column of ones to an array.
```

```python
model2 = sm.OLS(target , rm1)
fitted_model2 = model2.fit()

model3 = sm.OLS(target , lstat1)
fitted_model3 = model3.fit()
```

```python
# rm 결과 출력
fitted_model2.summary()
```

```python
# lstat 결과 출력
fitted_model3.summary()
```

```python
# 각 각 yhat 예측하기
predict2 = fitted_model2.predict(rm1)
predict3 = fitted_model3.predict(lstat1)
```

```python
import matplotlib.pyplot as plt
plt.yticks(fontname = 'Arial')
plt.scatter(rm, target, label = 'data')
plt.plot(rm , predict2, label = 'result')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.yticks(fontname = 'Arial')
plt.scatter(lstat, target, label = 'data')
plt.plot(lstat , predict3, label = 'result')
plt.legend()
plt.show()
```

```python
plt.scatter(target, predict2) 
plt.xlabel('real_value')
plt.ylabel('predict_value')
plt.show()
```

```python
plt.scatter(target, predict3) 
plt.xlabel('real_value')
plt.ylabel('predict_value')
plt.show()
```
