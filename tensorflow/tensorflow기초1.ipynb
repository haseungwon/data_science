```
import numpy as np
import tensorflow as tf
```

Tensor 생성

- tf.constant()
  - list -> Tensor
  - array - >Tensor

```
tf.constant([1,2,3])
tf.constant(((1,2,3,),(1,2,3)))
```

```
arr = np.array(np.random.randint(5, size = 10))
```

```
tensor = tf.constant(arr)
tensor
```

shape 확인

```
tensor.shape
```

data type 확인
  - 주의: tensor 생성 할 때도 data type을 정해주지 않기 때문에 data type에 대한 혼동이 올 수 있음
  - data type에 따라 모델의 무게나 성능 차이에도 영향을 줄 수 있음

```
tensor.dtype
```

data type 정의

```
tf.constant(arr, dtype=tf.float32)
```

```
arr = np.array([1,2,3], dtype=np.float32)
```

```
arr.astype(np.uint32)
```

난수 생성
  - normal distribution 은 중심극한 이론에 의한 연속적인 모양
  - uniform distribution은 중심 극한 이론과는 무고나하며 불연속적이며 일정한 분포
  

- numpy에서는 normal distribution을 기본적으로 생성
  - np.random.randn()

```
np.random.randn(3)
```

tf.random.noraml
  - tensorflow에서는 normal distribution

```
tf.random.normal([3,3])
```

tf.random.uniform
  - tensorflow에서는 uniform distribution

```
tf.random.uniform([3,3])
```
