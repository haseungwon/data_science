### crim/rm/lstat 3개의 변수로 각각 단순 선형 회귀 분석

```python
## 변수 설정 target/crim/rm/lstat
target = boston[['Target']]
crim = boston[['CRIM']]
rm = boston[['RM']]
lstat = boston[['LSTAT']]
```

### target ~ crim 선형회귀분석

```python
# crim변수에 상수항 추가하기 y = b0+b1x 이기때문에
crim1 = sm.add_constant(crim, has_constant = "add") # Add a column of ones to an array.
```

```python
crim1.head()
```

```python
# sm.OLS 적합시키기 ordinary least squares
model1 = sm.OLS(target , crim1)
fitted_model1 = model1.fit()
```

```python
fitted_model1.summary()
# R-squared: ssr/sst
```

```python
# 회귀 계수 출력

# fitted_model1.params 의 반환은 series이다

print( fitted_model1.params)
print( fitted_model1.params['CRIM'])
```

### y_hat = beta0 + beta1*x 계산하기

```python
print(crim1.shape)
print(fitted_model1.params.shape)
```

```python
# 회귀 계수 x 데이터(X)
# crim(1 , x) * fitted_model1.params(const, 회귀계수)
np.dot(crim1, fitted_model1.params)
```

```python
# predict 함수를 통해 yhat 구하기
predict = fitted_model1.predict(crim1)
```

```python
## 직접 구한 yhat과 predict 함수를 통해 구한 yhat 차이
predict - np.dot(crim1, fitted_model1.params)
```

### 적합시킨 직선 시각화

```python
import matplotlib.pyplot as plt
plt.yticks(fontname = 'Arial')
plt.scatter(crim, target, label = 'data')
plt.plot(crim , predict, label = 'result')
plt.legend()
plt.show()
```

```python
plt.scatter(target, predict) 
plt.xlabel('real_value')
plt.ylabel('predict_value')
plt.show()
# 잘 맞지 않음
```

```python
## residual 시각화 잔차

fitted_model1.resid.plot()
plt.xlabel('residual_number')
plt.show()
```

```python
# 잔차의 합계산하기 ->잔차는 0에 수렴할 것이다 
np.sum(fitted_model1.resid)
```

## 다중선형회귀

### 행렬연산을 통해 beta구하기


```python
from numpy import linalg # 행렬연산을 통해 beta구하기  (X'X)-1X'y
k = linalg.inv(np.dot(x_data1.T, x_data1))
np.dot(np.dot(k,x_data1.T),target)
```

```python
# y_hat 구하기
np.dot(x_data1,fitted_multi_model.params)
predict4 = fitted_multi_model.predict(x_data1)
predict4 - np.dot(x_data1,fitted_multi_model.params)
```

```python
# residual plot

fitted_multi_model.resid.plot()
plt.xlabel("residual plot")
plt.show()
```

### VIF를  통한 다중공선성 확인

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(x_data2.values, i) for i in range(x_data2.shape[1])]
vif['features'] = x_data2.columns
vif
```

```python
# nox 변수 제거후 (x_data3)  VIF 확인하기
vif = pd.DataFrame()
x_data3 = x_data2.drop('NOX', axis = 1)
vif['VIF Factor'] = [variance_inflation_factor(x_data3.values, i) for i in range(x_data3.shape[1])]
vif['features'] = x_data3.columns
vif
# 상관행렬 + p value에서 봤을땐 RM을 굳이 삭제할 필요는 없다
```

```python
# RM을 제거한 후 (x_data4) VIF 확인하기
vif = pd.DataFrame()
x_data4 = x_data3.drop('RM', axis = 1)
vif['VIF Factor'] = [variance_inflation_factor(x_data4.values, i) for i in range(x_data4.shape[1])]
vif['features'] = x_data4.columns
vif
```

```python
# nox 변수 제거한 데이터(x_data3) 상수항 추가후 회귀모델 적합
# nox, rm 변수 제거한 데이터(x_data4) 상수항 추가후 회귀모델 적합

x_data3 = sm.add_constant(x_data3, has_constant = 'add')
x_data4 = sm.add_constant(x_data4, has_constant = 'add')
model_vif = sm.OLS(target, x_data3)
fitted_model_vif = model_vif.fit()
model_vif2 = sm.OLS(target,  x_data4)
fitted_model_vif2 = model_vif2.fit()
```

```python
fitted_model_vif.summary()
# r*2은 사실 소수점 아래점에서는 더 작을것임 기존보다
# nox가 차지하는 변동성은 작았다
```

```python
fitted_model_vif2.summary()
# r*2 가 확실히 감소함 -> rm은 제거하면 안댐
# AIC는 낮을수록 좋음
```
